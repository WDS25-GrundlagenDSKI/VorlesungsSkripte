```python
import pandas as pd
import numpy as np
```

```python
# CSV-Datei laden
# https://www.kaggle.com/datasets/akashbommidi/super-market-sales/data
df = pd.read_csv('supermarket_sales.csv') # Dateipfad übergeben

# Erste 5 Zeilen anzeigen
df.head()
```

```python
# Anzeige der Spaltenbezeichnungen
print('Spalten: ', df.columns)

# Index - Zeilen haben typischerweise einen numerischen Index, können aber z.B. auch vom Typ String sein
print(df.index)
```

```python
# Übersicht und statistische Werte
df.describe()
```

```python
# Übersicht und statistische Werte von Variablen vom typ Object
df.describe(include=object)
```

```python
# Überblick über Datentypen und fehlende Werte
df.info()
```

```python
# Ausgabe aller Rechnungen mit mehr als 100$
df[df.Total>100] # Alterativ df[df['Total']>100]
```

```python
# Prüfen ob Unit Price*Quantity = Total? Die Summe darüber müsste 0 sein
print('Summe über  Total-Unit price*quantity: ', sum(df['Total']-df['Unit price']*df['Quantity']))

# Prüfen ob der Stückpreis noch ohne Steuer ist
print('Summe über Total-Unit price*1.05*quantity: ',sum(df['Total']-df['Unit price']*1.05*df['Quantity'])) # 5% tax auf den Unit price
```

```python
# Total nach City gruppieren, wie viel wird ingesamt je Stadt ausgegeben
total_per_city = df.groupby('City')['Total'].sum()
print(total_per_city)
print('\n') # Newline in der Ausgabe

# Wie hoch ist Total im Schnitt je Geschlecht
total_per_gender = df.groupby('Gender')['Total'].mean()
print(total_per_gender)
print('\n')

# Kategore nach Gender gruppieren
gender_product_line = df.groupby(['Product line', 'Gender']).size()
print(gender_product_line)
```

## Nominal skalierte Varialben

```python
from scipy.stats.contingency import association


# Kontingenztabelle erstellen und ausgeben
ct = pd.crosstab(df['Product line'], df['Gender'])
print(ct)
# Berechnen und Ausgabe von Cramers V
print('Cramers V zwischen Product Line und Gender: ', association(ct, method="cramer"))

ct = pd.crosstab(df['Gender'], df['Payment'])
print(ct)
print('Cramers V zwischen Payment und Gender: ', association(ct, method="cramer"))


ct = pd.crosstab(df['City'], df['Customer type'])
print(ct)
print('Cramers V zwischen City und Customer type: ', association(ct, method="cramer"))
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
ct = pd.crosstab(titanic['sex'], titanic['survived'])
print(ct)
print('Cramers V zwischen Geschlecht und survived: ', association(ct, method="cramer"))

ct = pd.crosstab(titanic['sex'], titanic['alone'])
print(ct)
print('Cramers V zwischen Geschlecht und alone: ', association(ct, method="cramer"))
```
