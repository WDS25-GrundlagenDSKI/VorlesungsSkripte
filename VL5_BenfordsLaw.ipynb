```python
import math
import numpy as np
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

import refurb
```

```python
# =============================================================================
# # Create Benfords Law
# =============================================================================

N = pd.DataFrame(range(1,10,1), columns=['x'])
display(N)

#N["Benford"] = np.log(N['x']+1)-np.log(N['x'])
N["Benford"] = np.log(1+1/N['x'])
print(N.columns)
display(N)

# =============================================================================
# # Read Staedte XLS File from 
# =============================================================================
staedte = pd.read_excel('/Users/schoch/Documents/0_Vorlesungen/Statistik/Vorlesung_JSC/01_Einfuehrung/05-staedte.xlsx', sheet_name=1, header=[4],skiprows=[5,6])
print(staedte.columns)

staedte = staedte[["Unnamed: 6", "Unnamed: 8", "insgesamt"]]
staedte.columns = ["Stadt", "Flaeche", "Bevoelkerung"]
print("Size before dropna: ", staedte.size)
staedte = staedte.dropna()
staedte["Bevoelkerung"] = staedte["Bevoelkerung"].astype(int)
display(staedte)

# Get first digit of Flaeche and Bevoelkerung
print("Size after dropna: ", staedte.size)
staedte["1_Flaeche"] = staedte["Flaeche"].astype(str).str[0].astype(int)
staedte["1_Bevoelkerung"] = staedte["Bevoelkerung"].astype(str).str[0].astype(int)
display(staedte) 



```

```python
# =============================================================================
# # Plots 
# =============================================================================
fig1 = make_subplots(rows=1, cols=3, subplot_titles=("log(N+1)-log(N)", "Verteilung", "Verteilung"))

# Scatterplot manuell hinzufügen
fig1.add_trace(
    go.Scatter(x=N["x"], y=N["Benford"], mode='markers'),
    row=1, col=1
)

fig1.add_trace(
    go.Histogram(x=staedte["1_Flaeche"], histnorm="probability density", name='Erste Ziffer Flaeche', xbins=dict(start=0, end=9, size=.5)),
    row=1, col=2
)
fig1.add_trace(
    go.Histogram(x=staedte["1_Bevoelkerung"], histnorm="probability density", name='Erste Ziffer Bevoelkerung', xbins=dict(start=0, end=9, size=.5)),
    row=1, col=3
)

fig1.update_layout(height=600, width=800, title_text="", xaxis_title_text="Ziffer", yaxis_title_text="")
fig1.update_xaxes(title_text="Ziffer", row=1, col=1)
fig1.update_xaxes(title_text="Ziffer", row=1, col=2)
fig1.update_xaxes(title_text="Ziffer", row=1, col=3)
fig1.show()

```
